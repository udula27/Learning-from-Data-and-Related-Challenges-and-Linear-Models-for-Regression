```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#Generate 100 samples

n_samples = 100

#Generating uniformly distributed values between 0 and 10

X = 10 * np.random.rand(n_samples, 1)

epsilon = np.random.normal(0,15, n_samples)

Y = 3 + 2 * X + epsilon[:, np.newaxis]
```

```python
#Data visualization

r = np.random.randint(104)

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=r)

#plot the data points

plt.figure(figsize = (10,6))

plt.scatter( X_train , Y_train , alpha =1 , marker ='o', color ='red', label ='Training Data ')
plt.scatter( X_test , Y_test , alpha =1 , marker ='s', color ='blue', label ='Testing Data ')
plt.show ()

```

```python
for i in range(10):
  X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state = np.random.randint(104))
  model = LinearRegression()
  model.fit(X_train, Y_train)
  Y_pred_train = model.predict(X_train)
  plt.plot(X_train, Y_pred_train, label = f'LR {i+1}')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
infrared_thermography_temperature = fetch_ucirepo(id=925)

# data (as pandas dataframes)
X = infrared_thermography_temperature.data.features
y = infrared_thermography_temperature.data.targets

# metadata
print(infrared_thermography_temperature.metadata)

# variable information
print(infrared_thermography_temperature.variables)
```

```python
import pandas as pd

data = pd.concat([X, y], axis = 1)

data = data.dropna()

X_clean = data.iloc[:, :-1]
y_clean = data.iloc[:, -1]
```

```python


df = pd.DataFrame(data)

# Function to convert age range to mean
def age_range_to_mean(age_range):
    if pd.isna(age_range):  # Check for NaN values
        return None  # Return None for NaN
    if age_range.startswith('>'):
        # Handle case for ">60"
        return 65
    elif age_range.startswith('<'):
        # Handle case for "<21"
        return 18
    else:
        # Split the string into lower and upper bounds
        lower, upper = map(int, age_range.split('-'))
        # Calculate the mean
        return (lower + upper) / 2

# Apply the function to the Age column
df['Age'] = df['Age'].apply(age_range_to_mean)

df = df.dropna(subset=['Age', 'aveOralM'])

X = df
y = df[['aveOralM','aveOralM']]
```

```python
import numpy as np
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error, r2_score

# 1. Residual sum of squares (RSS)
y_pred = model.predict(X_test)
rss = np.sum(np.square(y_test - y_pred))

# 2. Residual Standard Error (RSE)
N = X_test.shape[0]
d = X_test.shape[1]
rse = np.sqrt(rss / (N - d - 1))

# 3. Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# 4. R^2 statistic
r2 = r2_score(y_test, y_pred)

# 5. Standard error, t-statistic, and p-value for each feature
# Add a constant term to the independent variables for statsmodels
X_train_sm = sm.add_constant(X_train)
model_sm = sm.OLS(y_train, X_train_sm).fit()

# Standard errors, t-statistics, and p-values
standard_errors = model_sm.bse
t_values = model_sm.tvalues
p_values = model_sm.pvalues

# Print the results
print(f"Residual sum of squares (RSS): {rss}")
print(f"Residual Standard Error (RSE): {rse}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R^2 statistic: {r2}")
print("\nStandard Errors for each feature:\n", standard_errors)
print("\nt-statistics for each feature:\n", t_values)
print("\np-values for each feature:\n", p_values)
```
